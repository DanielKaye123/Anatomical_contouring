```python
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk

# Create a sample 3D bitmap
#bitmap_3d = np.random.randint(0, 2, size=(256, 256, 256), dtype=np.uint8)  # Replace with your 3D bitmap data

#img1 = sitk.ReadImage("/vol/bitbucket/djk18/rank_10_mc_20_patch_110_cervical_2m_iso/test/predictions/zzAMLART081_prob_maps.nii.gz") 
img1 = sitk.ReadImage("/vol/bitbucket/djk18/rank_10_mc_20_patch_110_cervical_2m_iso/test/predictions/zzAMLART081_prob_maps.nii.gz") 
img2 = sitk.ReadImage("/vol/bitbucket/djk18/cervical_Task7/test/predictions/zzAMLART081_prob_maps.nii.gz")
img3 = sitk.ReadImage("/vol/bitbucket/djk18/cervical_Task16_1050_full/test/predictions/zzAMLART081_prob_maps.nii.gz")
# for key in img.GetMetaDataKeys():
#     print(key + " : " + img.GetMetaData(key))
def display(img):
  prob_map = sitk.GetArrayFromImage(img) # Convert the SimpleITK image to a NumPy array

  
  #prob_map = np.clip(prob_map, 0, 0,5)
  # Select a slice to display
  slice_index = 90
  class_num = 1

  slice_map = prob_map[slice_index, :, :, class_num]
  slice_map = np.clip(slice_map, -1, 0.1)
  
  # Display the selected slice with grid lines
  fig, ax = plt.subplots()
  print(slice_map.shape)
  ax.imshow(slice_map, cmap='hot')

  # Add grid lines
  ax.grid(color='white', linewidth=0.5)

  # Add a patch (110, 110, 110) over the image
  patch_center = (90, 50)
  patch_size = (110, 110)
  rect = plt.Rectangle((patch_center[0] - patch_size[0] // 2, patch_center[1] - patch_size[1] // 2),
                      patch_size[0], patch_size[1], linewidth=2, edgecolor='white', facecolor='none')
  ax.add_patch(rect)

  # Customize the plot
  ax.set_aspect('equal')
  plt.axis('off')

  # Show the plot
  plt.show()

display(img1)
display(img2)
display(img3)
```
